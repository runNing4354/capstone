```python
import librosa 
import numpy as np
import pandas as pd
import os
import librosa.display
from matplotlib import pyplot as plt

path = "C:\\Users\\taemi\\capstone\\data\\1_default\\sounddata\\vdzrctqrhw.wav"
max = 0

audio, sample_rate = librosa.load(path, res_type='kaiser_fast') 
if len(audio)/sample_rate > 2:
    audio = audio[ : sample_rate*2]
S = librosa.feature.melspectrogram(audio, sr=sample_rate, n_mels=64, hop_length=512)
log_S = librosa.power_to_db(S, ref=np.max)
librosa.display.specshow(log_S, sr=sample_rate, x_axis='time')
plt.title('Mel spectrogram')
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.colorbar(format='%+2.0f dB')
```
