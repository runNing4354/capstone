```python
##tonow.ui

form_now = resource_path('setting.ui')
form_nowwindow = uic.loadUiType(form_now)[0]

class nowwindow(QDialog,QWidget,form_nowwindow):
    def __init__(self):
        super(nowwindow,self).__init__()
        self.initUi()
        
        self.radioButton.clicked.connect(self.groupboxRadFunction)
        self.radioButton_2.clicked.connect(self.groupboxRadFunction)
        self.radioButton_3.clicked.connect(self.groupboxRadFunction)
        self.radioButton_4.clicked.connect(self.groupboxRadFunction)
        self.radioButton_5.clicked.connect(self.groupboxRadFunction)
        
    def groupboxRadFunction(self) :
        if self.radioButton.isChecked() : 
            print('hi')
            
        elif self.radioButton_2.isChecked() : 
            print('hi')
            
        elif self.radioButton_3.isChecked() : 
            print("GroupBox_rad3 Checked")
            
        elif self.radioButton_4.isChecked() : 
            print("GroupBox_rad4 Checked")
        
        elif self.radioButton_5.isChecked() : 
            print("GroupBox_rad4 Checked")

            
        
    def initUi(self):
        self.setupUi(self)
        
    def back(self):
        self.close()
        
    def scream(self):
        print('')
        
    def horn(self):
        print('')
        
    def cry(self):
        print('')
        
    def siren(self):
        print('')
        
    def start(self):
        

        CHUNK = 1024
        FORMAT = pyaudio.paInt16
        CHANNELS = 1
        RATE = 44100
        RECORD_SECONDS = 4
        WAVE_OUTPUT_FILENAME = "test.wav"

        p = pyaudio.PyAudio()

        stream = p.open(format=FORMAT,channels=CHANNELS,rate=RATE,input=True,frames_per_buffer=CHUNK)

        print("* recording")

        frames = []

        for i in range(0, int(RATE / CHUNK * RECORD_SECONDS)):
            data = stream.read(CHUNK)
            frames.append(data)

        print("* done recording")

        stream.stop_stream()
        stream.close()
        p.terminate()

        wf = wave.open(WAVE_OUTPUT_FILENAME, 'wb')
        wf.setnchannels(CHANNELS)
        wf.setsampwidth(p.get_sample_size(FORMAT))
        wf.setframerate(RATE)
        wf.writeframes(b''.join(frames))
        wf.close()
        
        filename = 'test.wav'
        answer = print_prediction(filename)
        
        if(answer == 'car_horn'):
            self.label_6.setPixmap(QtGui.QPixmap("horn.png")) #image path
        elif(answer == 'Meow'):
            self.label_6.setPixmap(QtGui.QPixmap("meow.png")) #image path
        elif(answer == 'siren'):
            self.label_6.setPixmap(QtGui.QPixmap("siren.png")) #image path
        elif(answer == 'gun_shot'):
            self.label_6.setPixmap(QtGui.QPixmap("gun.png")) #image path
        elif(answer == 'Squeak'):
            self.label_6.setPixmap(QtGui.QPixmap("squeak.png")) #image path
        elif(answer == 'Knock'):
            self.label_6.setPixmap(QtGui.QPixmap("knock.png")) #image path
        elif(answer == 'dog_bark'):
            self.label_6.setPixmap(QtGui.QPixmap("dog.png")) #image path
            
           
        
        
        
```

```python
max_pad_len = 1287
num_rows = 40
num_columns = 1287
num_channels = 1

model2=load_model('Saved Model/Weight_best5.hdf5')
def extract_features(file_name):
    try:
        audio, sample_rate = librosa.load(file_name, res_type = 'kaiser_fast')
        mfccs = librosa.feature.mfcc(y=audio, sr = sample_rate, n_mfcc = 40)
        pad_width = max_pad_len - mfccs.shape[1]
        mfccs = np.pad(mfccs, pad_width = ((0, 0), (0, pad_width)), mode = 'constant')
    
    except Exception as e:
        print("Error 발생: ", file_name)
        return None
    
    return mfccs

def print_prediction(file_name):
    prediction_feature = extract_features(file_name) 
    prediction_feature = prediction_feature.reshape(1, num_rows, num_columns, num_channels)
    y_prob = model2.predict(prediction_feature)
    predicted_vector = y_prob.argmax(axis=-1)
    predicted_class = le.inverse_transform(predicted_vector) 
    print("The predicted class is:", predicted_class[0], '\n') 


    predicted_proba_vector = model2.predict(prediction_feature) 
    predicted_proba = predicted_proba_vector[0]
    for i in range(len(predicted_proba)): 
        category = le.inverse_transform(np.array([i]))
        print(category[0], "\t\t : ", format(predicted_proba[i], '.32f') )
        
    return predicted_class
        
def predict_prob(number):
    return [number[0],1-number[0]]
```
